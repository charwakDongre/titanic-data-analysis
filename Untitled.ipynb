```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import StandardScaler 
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
df.head()
```

```python

df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
if 'Cabin' in df.columns:
    df.drop('Cabin', axis=1, inplace=True)

df.info()
```

```python

df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
df.drop(['Name', 'Ticket'], axis=1, inplace=True)
df.head()
```

```python

scaler = StandardScaler()
numerical_features = ['Age', 'Fare', 'SibSp', 'Parch', 'Pclass']
df[numerical_features] = scaler.fit_transform(df[numerical_features])
df.head()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[['Age', 'Fare']])
plt.title('boxplot for age and fare outlier removal')
plt.show()
```

```python
Q1 = df[['Age', 'Fare']].quantile(0.25)
Q3 = df[['Age', 'Fare']].quantile(0.75)
IQR=Q3-Q1
outlier_condition = ~((df[['Age', 'Fare']] < (Q1 - 1.5 * IQR)) | (df[['Age', 'Fare']] > (Q3 + 1.5 * IQR))).any(axis=1)
df_no_outliers = df[outlier_condition].copy()
print(df.shape)
print(df_no_outliers.shape)
```

```python
#After removing the outliers 
plt.figure(figsize=(12,6))
sns.boxplot(data=df_no_outliers[['Age', 'Fare']])
plt.title('boxplot after outlier removal of Age and Fare')
plt.show()
```

```python

df_no_outliers.to_csv('Titanic-Dataset-Cleaned.csv', index=False)
```
